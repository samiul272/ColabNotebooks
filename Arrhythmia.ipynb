```
###################---------------------  Gathering the data --------------------###############################

import wfdb
import os
import numpy

path = os.getcwd()+ r'/sampleData'

attributes = []
headers = []
data = []

for(dirpath,dirnames,filenames) in os.walk(path):
  for filename in filenames:
    if (filename.endswith('.hea')):
      headers.append(filename)
    elif(filename.endswith('.atr')):
      attributes.append(filename)
    elif(filename.endswith('.dat')):
      data.append(filename)
headers.sort()
attributes.sort()
data.sort()

#print(headers)
#print(attributes)
#print(data)


FirstColumn = []
SecondColumn = []


for dataPoint in data:
  record_no = dataPoint[0:3]
  record = wfdb.rdrecord(path+r'/'+record_no)
  wfdb.plot_wfdb(record = record,title = 'RecordForSampleNo_'+str(record_no))
  pSig_Dict = record.__dict__['p_signal']
  print('Record_No: ' + record_no)
  print(pSig_Dict)
  print('--------------------')
  FirstColumn.append(pSig_Dict[:,0])
  SecondColumn.append(pSig_Dict[:,1])
  
print(FirstColumn)
print(SecondColumn)
```

```
###################------------------- finding only_MLII containing sample number--------------------############################
only_MLII_samp_no =[]

for dataPoint in data:
  record_no = dataPoint[0:3]
  record = wfdb.rdrecord(path+r'/'+record_no)
  sig_Dict = record.__dict__['sig_name']
  print('Record_no: ' + record_no)
  print(sig_Dict)
  print('--------------------')
  if(sig_Dict[0] == 'MLII' ):
    only_MLII_samp_no.append(record_no)
  elif(sig_Dict[1] == 'MLII'):
    only_MLII_samp_no.append(record_no)
else:
  print('All Samples that have MLII :: ')
  print(only_MLII_samp_no)
  
```

```
#######################-------------Finding only MLII_containing sample no. and respective data-----------#########################################

only_MLII_samp_no =[]
only_MLII_Data = []

for dataPoint in data:
  record_no = dataPoint[0:3]
  record = wfdb.rdrecord(path+r'/'+record_no)
  sig_Dict = record.__dict__['sig_name']
  pSig_Dict = record.__dict__['p_signal']
  #print('Record_no: ' + record_no)
  #print(sig_Dict)
  #print('--------------------')
  if(sig_Dict[0] == 'MLII' ):
    only_MLII_samp_no.append(record_no)
    only_MLII_Data.append(pSig_Dict[:,0])
  elif(sig_Dict[1] == 'MLII'):
    only_MLII_samp_no.append(record_no)
    only_MLII_Data.append(pSig_Dict[:,1])
else:
  print('All Samples that have MLII :: ')
  print(only_MLII_samp_no)
  print(only_MLII_Data)
  
  
  
  
  

    
```
